```python
print("--- Soal 2 (Modifikasi dengan NIM) ---")
kode_berhenti = 4232411054
input_pengguna = 0  # Inisialisasi awal

while input_pengguna != kode_berhenti:
    try:
        input_pengguna = int(input("Masukkan angka (atau NIM anda untuk berhenti): "))
    except ValueError:
        print("Input tidak valid. Harap masukkan angka.")

print("Perulangan telah dihentikan oleh NIM anda (4232411054).")

```

```python
print("--- Soal 3 (Modifikasi) ---")
harga_barang = [15000, 25000, 10000, 5000, 12000, 8000, 20000, 30000, 18000, 22000]
total_belanja = 0

for harga in harga_barang:
    total_belanja += harga

print("Total biaya belanja adalah:", total_belanja)
```

```python
print("--- Soal 4 (Modifikasi dengan Password) ---")
percobaan_maksimal = 3
percobaan_saat_ini = 1
password_benar = "pedro1804"

while percobaan_saat_ini <= percobaan_maksimal:
    password = input(f"Percobaan ke-{percobaan_saat_ini}. Masukkan password: ")
    if password == password_benar:
        print("Akses diterima.")
        break
    else:
        print("Password salah.")
    percobaan_saat_ini += 1

if percobaan_saat_ini > percobaan_maksimal:
    print("Kesempatan habis.")

```

```python
print("--- Soal 5 ---")
nama_proyek = "Politeknik Negeri Batam"
hasil_output = ""

for karakter in nama_proyek:
    hasil_output += karakter + "-"

print(hasil_output)
# Output: P-o-l-i-t-e-k-n-i-k- -N-e-g-e-r-i- -B-a-t-a-m-
```

```python
print("--- Soal 6 ---")
# Countdown dari 255 ke 1
for i in range(255, 0, -1):
    print(i)
```

```python
print("--- Soal 7 ---")
kelurahan = ["Baloi Permai", "Tiban Baru", "Batu Selicin", "Buliang", "Telaga Punggur"]

for i, nama in enumerate(kelurahan):
    print(f"Kelurahan ke-{i + 1}: {nama}")
```

```python
print("--- Soal 8 ---")
data_angka = [10, 20, 30, 0, 50, 60]

for angka in data_angka:
    if angka == 0:
        break
    print(angka)

print("Program berhenti karena angka 0 ditemukan.")
```

```python
print("--- Soal 9 ---")
total = 0
bilangan = 1
batas = 20

while total <= batas:
    total += bilangan
    bilangan += 1

print("Total terakhir:", total)
```
