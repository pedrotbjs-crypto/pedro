```python
print("="*30)
print("SELAMAT DATANG DI KAFE GEMINI")
print("="*30)

print("Menu Makanan & Minuman:")
print("1. Kopi Hitam  - Rp 12.000")
print("2. Teh Manis   - Rp 8.000")
print("3. Roti Bakar  - Rp 15.000")
print("4. Mie Instan  - Rp 10.000")

pilihan = int(input("Masukkan nomor pilihan Anda (1-4): "))
jumlah = int(input("Masukkan jumlah pesanan: "))

# Menentukan harga berdasarkan pilihan
if pilihan == 1:
    nama_pesanan = "Kopi Hitam"
    harga = 12000
elif pilihan == 2:
    nama_pesanan = "Teh Manis"
    harga = 8000
elif pilihan == 3:
    nama_pesanan = "Roti Bakar"
    harga = 15000
elif pilihan == 4:
    nama_pesanan = "Mie Instan"
    harga = 10000
else:
    print("Pilihan tidak valid.")
    exit()

total = harga * jumlah

# Menentukan diskon berdasarkan total belanja
if total >= 50000:
    diskon = 0.7   # 70%
elif total >= 30000:
    diskon = 0.5   # 50%
elif total >= 20000:
    diskon = 0.3   # 30%
else:
    diskon = 0     # Tidak ada diskon

potongan = total * diskon
total_setelah_diskon = total - potongan

print("\n===============================")
print("Pesanan Anda :", nama_pesanan)
print("Jumlah       :", jumlah)
print(f"Total Harga  : Rp {total}")
print(f"Diskon       : {int(diskon*100)}%")
print(f"Total Bayar  : Rp {int(total_setelah_diskon)}")
print("===============================")

uang_str = input("Masukkan jumlah uang Anda: Rp ")
uang = int(uang_str.replace("Rp ", "").replace(".", "")) # Remove "Rp " and "." before converting to int

if uang >= total_setelah_diskon:
    kembali = uang - total_setelah_diskon
    print(f"Kembalian Anda: Rp {int(kembali)}")
    print("Terima kasih telah berbelanja di Kafe Gemini!")
else:
    print(f"Uang Anda kurang Rp {int(total_setelah_diskon - uang)}")
```
