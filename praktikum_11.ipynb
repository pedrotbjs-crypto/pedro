```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 7, 14, 20, 18]

# Membuat grafik
plt.plot(x, y)
plt.title("Grafik Garis Penjualan")
plt.xlabel("Hari")
plt.ylabel("Jumlah Penjualan")

# Menampilkan grafik
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
kategori = ['A', 'B', 'C', 'D']
nilai = [15, 25, 10, 20]

# Membuat grafik batang
plt.bar(kategori, nilai)
plt.title("Grafik Batang Nilai Kategori")
plt.xlabel("Kategori")
plt.ylabel("Nilai")

# Menampilkan grafik
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [5, 10, 15, 20, 25]
y = [7, 14, 20, 25, 30]

# Membuat scatter plot
plt.scatter(x, y)
plt.title("Scatter Plot Hubungan X dan Y")
plt.xlabel("Variabel X")
plt.ylabel("Variabel Y")

# Menampilkan grafik
plt.show()
```
